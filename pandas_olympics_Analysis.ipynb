#1. Load the Dataset

```python
import pandas as pd  #importing pandas package
df = pd.read_csv("/content/athlete_events.csv") #loading dataframe.
```

```python
#Viewing DataFrame
df.head()
```

#2.Data Exploration

```python
# Display the dataset information
df.info()
```

```python
# Display basic statistics of the dataset
df.describe()
```

#3.Data Cleaning

```python
# Check for missing values
missing_values = df.isnull().sum()
```

```python
# Display columns with missing values
print(missing_values[missing_values > 0])
```

```python
age_mean=df['Age'].mean()
height_mean=df['Height'].mean()
weight_mean=df['Weight'].mean()
```

```python
# Fill or drop missing values as needed (for simplicity, we'll drop rows with missing values)
df['Age'].fillna(value=age_mean,inplace=True)
df['Height'].fillna(value=height_mean,inplace=True)
df['Weight'].fillna(value=height_mean,inplace=True)
df['Medal'].fillna(value='no_medal',inplace=True)
```

```python
# Check for missing values
missing_values = df.isnull().sum()
print(missing_values[missing_values > 0])
```

```python
df_cleaned = df.copy() #giving a copy of cleaned DataFrame.
```

```python
df_cleaned.head()
```

#4. Data Analysis

**a. Total Medals by Country**

```python
# Group by country and count the total number of medals
medals_by_country = df_cleaned.groupby('Team')['Medal'].count().reset_index()

# Sort the results in descending order
medals_by_country = medals_by_country.sort_values(by='Medal', ascending=False)

# Display the top 10 countries with the most medals
print(medals_by_country.head(10))
```

```python
medals_by_country.shape[0]
```

```python
#I want to medals by India,China,Pakisthan,Bangladesh,Srilanka,Nepal
Neighbour_countries=["India","China","Pakistan","Bangladesh","Sri Lanka","Nepal"]
for i in Neighbour_countries:
  print(medals_by_country[medals_by_country['Team']==i])
```

```python
# how many gold , silver , bronge medals won by India
medals_by_country_years = df_cleaned.groupby(['Team','Medal']).count().reset_index()

medals_by_country_years[medals_by_country_years["Team"]=='India']
```

```python
ind_medals_till_2016=df_cleaned[df_cleaned.Team == 'India']['Medal'].value_counts()
ind_medals_till_2016
```

**b. Medals Over Time**

```python
# Group by year and count the total number of medals
medals_by_year = df_cleaned.groupby('Year')['Medal'].count().reset_index()

# Display the trend of medals over the years
print(medals_by_year)
```

**c. Medals by Sport**

```python
# Group by sport and count the total number of medals
medals_by_sport = df_cleaned.groupby(['Sport','Team'])['Medal'].count().reset_index()

# Sort the results in descending order
medals_by_sport = medals_by_sport.sort_values(by='Medal', ascending=False)

# Display the top 10 sports with the most medals
print(medals_by_sport.head(10))
```

```python
#I want to medals by India,China,Pakisthan,Bangladesh,Srilanka,Nepal
for i in Neighbour_countries:
  print(medals_by_sport[medals_by_sport['Team']==i])
```

**d. Medals by Gender**

```python
# Group by gender and count the total number of medals
medals_by_gender = df_cleaned.groupby(['Sex','Team'])['Medal'].count().reset_index()

# Display the medal counts by gender
print(medals_by_gender.head(10))
```

```python
for i in Neighbour_countries:
  print(medals_by_gender[medals_by_gender['Team']==i])
```
